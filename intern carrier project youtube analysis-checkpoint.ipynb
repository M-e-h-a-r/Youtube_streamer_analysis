```python
average_values = df[['Suscribers', 'Visits', 'Likes', 'Comments']].mean()
print("Average values:\n", average_values)
```

```python
above_average_streamers = df[(df['Suscribers'] > average_values['Suscribers']) & 
                             (df['Visits'] > average_values['Visits']) & 
                             (df['Likes'] > average_values['Likes']) & 
                             (df['Comments'] > average_values['Comments'])]

print("Streamers with above-average performance:\n", above_average_streamers)
```

```python
top_performers = above_average_streamers.sort_values(by='Suscribers', ascending=False)

print("Top-performing content creators:\n", top_performers.head(10))  
```

## Content Recommendations:


```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import StandardScaler, OneHotEncoder

data = pd.get_dummies(df, columns=['Categories'])

# Define features and target variable
numerical_features = ['Suscribers', 'Likes', 'Comments', 'Visits']
categorical_features = [col for col in data.columns if col.startswith('Categories_')]
features = numerical_features + categorical_features
target = 'Brand_Collabs'

# Normalize the features
scaler = StandardScaler()
data[features] = scaler.fit_transform(data[features])

# Calculate similarity matrix based on features
similarity_matrix = cosine_similarity(data[features])

# Function to recommend content based on user preferences
def recommend_content(user_preferences, similarity_matrix, data):
    # Find the index of the most similar streamer to the user preferences
    most_similar_index = similarity_matrix[user_preferences].argsort()[-2]
    
    # Retrieve recommended content based on the most similar streamer
    recommended_content = data.loc[data.index == most_similar_index, 'Links'].values[0]
    
    return recommended_content

# Example usage:
# Let's say the user prefers streamer with index 0
user_preferences = 0
recommended_content = recommend_content(user_preferences, similarity_matrix, data)
print("Recommended content:", recommended_content)

```
